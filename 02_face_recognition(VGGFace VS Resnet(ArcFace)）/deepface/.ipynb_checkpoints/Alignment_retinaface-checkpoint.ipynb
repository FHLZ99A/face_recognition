```python
from retinaface import RetinaFace
import math
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img_path = "tests/dataset/img11.jpg"
```

```python
def verify(img1_path,model_name = "ArcFace"):
    img1 = cv2.imread(img_path)
    
    
    plt.imshow(img1[:, :, :: -1])
    plt.show()
```

```python
verify(img_path)
```

```python
img = cv2.imread(img_path)
```

```python
resp = RetinaFace.detect_faces(img_path = img_path)
```

```python
resp
```

```python
x1, y1 = resp["face_1"]["landmarks"]["right_eye"]
x2, y2 = resp["face_1"]["landmarks"]["left_eye"]
```

```python
a = abs(y1 - y2)
b = abs(x1 - x2)
c = math.sqrt(a*a + b*b)
```

```python
a,b,c
```

```python
cos_alpha = (b*b + c*c - a*a)/(2*b*c)
```

```python
cos_alpha
```

```python
alpha = np.arccos(cos_alpha) #半径
```

```python
alpha = (alpha * 180)/math.pi
```

```python
alpha
```

```python
aligned_img = Image.fromarray(img)
```

```python
aligned_img = np.array(aligned_img.rotate(alpha))
```

```python
plt.imshow(aligned_img[:,:,::-1])
```
