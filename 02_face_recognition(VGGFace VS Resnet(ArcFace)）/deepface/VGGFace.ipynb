```python
#!pip install deepface
from deepface import DeepFace
import matplotlib.pyplot as plt
import pandas as pd
```

```python
img1_path = 'tests/dataset/img2.jpg'
img2_path = 'tests/dataset/img6.jpg'
```

```python
img1 = DeepFace.detectFace(img1_path)
img2 = DeepFace.detectFace(img2_path)
```

```python

plt.figure()
plt.subplot(1,2,1)
plt.imshow(img1)
plt.subplot(1,2,2)
plt.imshow(img2)
plt.show()

resp = DeepFace.verify(img1_path=img1_path, img2_path=img2_path)

resp
```
