```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
def verify(img1_path, img2_path, model_name = "VGG-Face"):
    img1 = cv2.imread(img1_path)
    img2 = cv2.imread(img2_path)
    
    plt.imshow(img1[:, :, :: -1])
    plt.show()
    plt.imshow(img2[:, :, :: -1])
    plt.show()
    
    result = DeepFace.verify(img1_path, img2_path, model_name=model_name)
    print("result:", result)
    
    verification = result['verified']
    if verification:
        print("they are same!")
    else:
        print("they are not same!")
```

```python
verify("img1.jpg", "img2.jpg")
```

```python
verify("img1.jpg", "img2.jpg", model_name = "Facenet")
```

```python
verify("img5.jpg", "img6.jpg", model_name = "OpenFace")
```

```python
verify('img1.jpg', 'img3.jpg')
```
